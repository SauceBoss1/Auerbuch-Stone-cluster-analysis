# Cluster Analysis based on TMM data

```python
from clusterTMM import ClusterTMM
import sklearn.preprocessing as skp
```

#### Setting up files and initializing pipelines

```python
coldData = 'master means/26C_mean_masterFile.tsv'
warmData = 'master means/37C_mean_masterFile.tsv'
names = ['pil', 'pcnB', 'pil-pcnB', 'WT']

ctmm1 = ClusterTMM(file=coldData, scaler=skp.Normalizer(), names=names)
ctmm2 = ClusterTMM(file=warmData, scaler=skp.Normalizer(), names=names)
```

#### Executing cluster analysis for 26C data

```python
ctmm1.plotElbow()
```

```python
df1 = ctmm1.cluster(n_clusters=3, showCenters=False, showDist=True)
df1.to_csv('cluster data/26C_clusters.tsv', sep='\t')
```

#### Executing Cluster Analysis for 37C data

```python
ctmm2.plotElbow()
```

```python
df = ctmm2.cluster(n_clusters=3, showDist=True)
df.to_csv('cluster data/37C_clusters.tsv', sep='\t')
```

### Creating heatmaps

```python
#26C
ctmm1.createHeatmap(title='26C', export=True, outName='heatmaps/26C_w_normalizer.pdf')
```

```python
#37C
ctmm2.createHeatmap(title='37C', export=True, outName='heatmaps/37C_w_normalizer.pdf')
```

### Complete Cluster Analysis with pil and pil-pcnB data dropped

```python
colsToDrop = ['pil','pil-pcnB']
ctmm1.dropCols(columns=colsToDrop)
ctmm2.dropCols(columns=colsToDrop)
```

```python
ctmm1.plotElbow()
```

```python
ctmm2.plotElbow()
```

```python
df1 = ctmm1.cluster(n_clusters=3, showDist = False)
df2 = ctmm2.cluster(n_clusters=3, showDist = False)
df1.to_csv('cluster data/26C_clusters_dropped.tsv',sep='\t')
df2.to_csv('cluster data/37C_clusters_dropped.tsv', sep='\t')
```

```python
#26C heatmap
ctmm1.createHeatmap(title='26C with pil and pil/pcnB dropped', export=True, outName='heatmaps/26C_w_normalizer_dropped.pdf')

#37C heatmap
ctmm2.createHeatmap(title='37C with pil and pil/pcnB dropped', export=True, outName='heatmaps/37C_w_normalizer_dropped.pdf')
```
